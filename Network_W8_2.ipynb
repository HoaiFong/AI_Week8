```python
import pandas as pd
url = 'crabs.csv'
crabs = pd.read_csv(url)
crabs.head(30)
```

```python

data = crabs.replace(["F","M","B","O"],[1,0,1,0])
data
```

```python
from sklearn.linear_model import Perceptron
from sklearn import datasets
from sklearn.model_selection import train_test_split # chia ra 7 phan train, 3 phan test
from sklearn.metrics import accuracy_score
import numpy as np

from sklearn.preprocessing import StandardScaler
#scale các dữ liệu không đều theo 1 tỉ lệ từ 0 -> 1


X1 = data['fontallip'].values
X2 = data['sex'].values
X3 = data['index'].values
X4 = data['rearwidth'].values
X5 = data['length'].values
X6 = data['width'].values
X7 = data['depth'].values

y = data['speies'].values

col_X1 = X1.reshape(-1,1)
col_X2 = X2.reshape(-1,1)
col_X3 = X3.reshape(-1,1)
col_X4 = X4.reshape(-1,1)
col_X5 = X5.reshape(-1,1)
col_X6 = X6.reshape(-1,1)
col_X7 = X7.reshape(-1,1)

X = np.hstack((col_X1,col_X2,col_X3,col_X4,col_X5,col_X6,col_X7))

print("Number of columns:", col_X1.shape[0])

print(X[:7][:7])

print(y[:20])
```

```python
X_train, X_test,y_train, y_test = train_test_split(X,y,test_size = 0.3) #test =0.3, train = 0.7

sc =  StandardScaler()
sc.fit(X)

X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)


print(X_train_std[:5],X_train[:5])

print(y[:5])
```

```python
model = Perceptron(max_iter= 40, eta0 = 0.1, random_state = 0)
# số lần học: max_iter 
# eta0 :  tốc độ học  giá trị từ 0 đến 1
# random_state : trạng thái giá trị ngẫu nhiên

model.fit(X_train_std, y_train)
y_pred = model.predict(X_test_std)
print(y_pred)
print(y_test)
print('Độ chính xác mô hình: ', accuracy_score(y_test, y_pred))
```

```python
model = Perceptron(max_iter= 1000, eta0 = 0.3, random_state = 0)
# số lần học: max_iter 
# eta0 :  tốc độ học  giá trị từ 0 đến 1
# random_state : trạng thái giá trị ngẫu nhiên

model.fit(X_train_std, y_train)
y_pred = model.predict(X_test_std)
print(y_pred)
print(y_test)
print('Độ chính xác mô hình: ', accuracy_score(y_test, y_pred))
```
